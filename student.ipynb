```python
class Student:

    def __init__(self, name, rollNumber):
        self.name = name
        self.rollNumber = rollNumber

    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.rollNumber)

class UndergraduateStudent(Student):

    def __init__(self, name, rollNumber, m1, m2, m3):
        super().__init__(name, rollNumber)
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def calculateAverage(self):
        average = (self.m1 + self.m2 + self.m3) / 3
        return average

class PostgraduateStudent(Student):

    def __init__(self, name, rollNumber, m1, m2, m3, m4):
        super().__init__(name, rollNumber)
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
        self.m4 = m4

    def calculateAverage(self):
        average = (self.m1 + self.m2 + self.m3 + self.m4) / 4
        return average

def main():

    ug = UndergraduateStudent("Kaviya", 101, 80, 75, 85)
    print("\nUndergraduate Student Details")
    ug.display()
    print("Average Marks:", ug.calculateAverage())

    # Postgraduate Student Object
    pg = PostgraduateStudent("Arun", 202, 70, 88, 90, 85)
    print("\nPostgraduate Student Details")
    pg.display()
    print("Average Marks:", pg.calculateAverage())

if __name__ == "__main__":
    main()
```

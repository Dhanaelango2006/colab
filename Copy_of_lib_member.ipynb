#include <iostream>
using namespace std;

// Base Class
class Member {
protected:
    string name;
    int membershipID;

public:
    Member(string n, int id) {
        name = n;
        membershipID = id;
    }

    virtual void display() {
        cout << "Name: " << name << endl;
        cout << "Membership ID: " << membershipID << endl;
    }

    virtual int getBorrowLimit() = 0; // Pure virtual function
};

// Derived Class - RegularMember
class RegularMember : public Member {
public:
    RegularMember(string n, int id) : Member(n, id) {}

    int getBorrowLimit() override {
        return 3; // Borrowing limit = 3 books
    }

    void display() override {
        cout << "Member Type: Regular Member" << endl;
        cout << "Name: " << name << endl;
        cout << "Borrowing Limit: " << getBorrowLimit() << " books" << endl;
        cout << "-----------------------------" << endl;
    }
};

// Derived Class - PremiumMember
class PremiumMember : public Member {
public:
    PremiumMember(string n, int id) : Member(n, id) {}

    int getBorrowLimit() override {
        return 10; // Borrowing limit = 10 books
    }

    void display() override {
        cout << "Member Type: Premium Member" << endl;
        cout << "Name: " << name << endl;
        cout << "Borrowing Limit: " << getBorrowLimit() << " books" << endl;
        cout << "-----------------------------" << endl;
    }
};

int main() {
    // Create one RegularMember object
    RegularMember r1("Arun", 101);

    // Create one PremiumMember object
    PremiumMember p1("Meena", 202);

    // Display their details
    r1.display();
    p1.display();

    return 0;
}

````python
%%writefile main.cpp
#include <iostream>
#include <string>
using namespace std;

// Base Class
class Member {
protected:
    string name;
    int membershipID;

public:
    Member(string n, int id) {
        name = n;
        membershipID = id;
    }

    virtual void display() {
        cout << "Name: " << name << endl;
        cout << "Membership ID: " << membershipID << endl;
    }

    virtual int getBorrowLimit() = 0; // Pure virtual function
};

// Derived Class - RegularMember
class RegularMember : public Member {
public:
    RegularMember(string n, int id) : Member(n, id) {}

    int getBorrowLimit() override {
        return 3; // Borrowing limit = 3 books
    }

    void display() override {
        cout << "Member Type: Regular Member" << endl;
        cout << "Name: " << name << endl;
        cout << "Borrowing Limit: " << getBorrowLimit() << " books" << endl;
        cout << "-----------------------------" << endl;
    }
};

// Derived Class - PremiumMember
class PremiumMember : public Member {
public:
    PremiumMember(string n, int id) : Member(n, id) {}

    int getBorrowLimit() override {
        return 10; // Borrowing limit = 10 books
    }

    void display() override {
        cout << "Member Type: Premium Member" << endl;
        cout << "Name: " << name << endl;
        cout << "Borrowing Limit: " << getBorrowLimit() << " books" << endl;
        cout << "-----------------------------" << endl;
    }
};

int main() {
    // Create one RegularMember object
    RegularMember r1("Dhana", 101);

    // Create one PremiumMember object
    PremiumMember p1("Sarish", 202);

    // Display their details
    r1.display();
    p1.display();

    return 0;
}
````
